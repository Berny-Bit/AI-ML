### 1.2 Handling Missing Data & Basic Statistics
In order to deal with the missing data, they get replaced with median value. That way, its robust to outliers. if we would have picked the average, and a certain value was inflated because of a charging session that artificially took a long time, it would effect the average when its an outlier. So imputation is used. imputation is when the missing value is replaced with the median value.

```python
print("Missing values  before imputation")
print(df.isnull().sum())

numeric_imputer = SimpleImputer(strategy='median')
cols_with_missing = ['Energy Consumed (kWh)',
    'Charging Rate (kW)',
    'Distance Driven (since last charge) (km)']

df[cols_with_missing] = numeric_imputer.fit_transform(df[cols_with_missing])
```
