### Phase 1: Data Understanding & Preparation
#### Step 1.1: Initial Data Exploration

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.impute import SimpleImputer
import warnings
warnings.filterwarnings('ignore')

# Advanced Python: Dictionary comprehension for settings
plot_settings = {
    'figure.figsize': (12, 8),
    'axes.titlesize': 14,
    'axes.labelsize': 12
}
plt.rcParams.update(plot_settings)

# Load the dataset (you'll need to replace this with your actual data loading)
df = pd.read_csv('ev_charging_patterns.csv')

# For now, let's assume we have the data loaded as df
print("Dataset Overview:")
print(f"Shape: {df.shape}")
print("\nData Types:")
print(df.dtypes)
print("\nMissing Values:")
print(df.isnull().sum())
```

#### Step 1.2: Handling Missing Data & Basic Statistics

```python
# Handle missing values using multiple strategies
print("Missing values before handling:")
print(df.isnull().sum())
missing_before = df.isnull().sum()

# Impute missing values - demonstrating different strategies
numeric_imputer = SimpleImputer(strategy='median')
columns_with_missing = ['Energy Consumed (kWh)', 'Charging Rate (kW)', 'Distance Driven (since last charge) (km)']

df[columns_with_missing] = numeric_imputer.fit_transform(df[columns_with_missing])

print("\nMissing values after handling:")
print(df.isnull().sum())

# Basic statistical summary
print("\nNumerical Features Summary:")
print(df.describe())

# Categorical features analysis
print("\nCategorical Features Unique Values:")
categorical_cols = ['Vehicle Model', 'Charging Station Location', 'Time of Day', 
                   'Day of Week', 'Charger Type', 'User Type']
for col in categorical_cols:
    print(f"{col}: {df[col].nunique()} unique values")
```

#### Step 1.3: Feature Engineering

```python
# Create new features from existing data
# Charging efficiency feature
df['Charging_Efficiency'] = (df['State of Charge (End %)'] - df['State of Charge (Start %)']) / df['Energy Consumed (kWh)']

# Time-based features from datetime
df['Charging Start Time'] = pd.to_datetime(df['Charging Start Time'])
df['Charging_Start_Hour'] = df['Charging Start Time'].dt.hour
df['Charging_Start_Month'] = df['Charging Start Time'].dt.month

# Battery usage pattern
df['Battery_Usage_Rate'] = df['Distance Driven (since last charge) (km)'] / (df['State of Charge (Start %)'] - df['State of Charge (End %)']).replace(0, np.nan)
df['Battery_Usage_Rate'].fillna(df['Battery_Usage_Rate'].median(), inplace=True)

# Cost efficiency
df['Cost_Per_kWh'] = df['Charging Cost (USD)'] / df['Energy Consumed (kWh)']

print("New features created:")
print(df[['Charging_Efficiency', 'Charging_Start_Hour', 'Battery_Usage_Rate', 'Cost_Per_kWh']].describe())
```

### Phase 2: Statistical Analysis & Data Preprocessing
#### Step 2.1: Statistical Testing with T-tests and ANOVA

```python
from scipy.stats import ttest_ind, f_oneway
import scipy.stats as stats

print("=" * 60)
print("STATISTICAL ANALYSIS FOR FEATURE SELECTION")
print("=" * 60)

# T-test: Compare energy consumption between different user types
print("\n1. T-TEST: Energy Consumption by User Type")
user_types = df['User Type'].unique()
for i in range(len(user_types)):
    for j in range(i+1, len(user_types)):
        group1 = df[df['User Type'] == user_types[i]]['Energy Consumed (kWh)']
        group2 = df[df['User Type'] == user_types[j]]['Energy Consumed (kWh)']
        t_stat, p_value = ttest_ind(group1, group2, equal_var=False)
        print(f"   {user_types[i]} vs {user_types[j]}: t-stat = {t_stat:.3f}, p-value = {p_value:.4f}")

# ANOVA: Test if charging efficiency differs by charger type
print("\n2. ANOVA: Charging Efficiency by Charger Type")
charger_groups = [df[df['Charger Type'] == charger]['Charging_Efficiency'] 
                  for charger in df['Charger Type'].unique()]
f_stat, p_value = f_oneway(*charger_groups)
print(f"   F-statistic = {f_stat:.3f}, p-value = {p_value:.4f}")

# Test for normal distribution using Shapiro-Wilk test
print("\n3. NORMALITY TESTS (Shapiro-Wilk):")
for feature in ['Energy Consumed (kWh)', 'Charging Duration (hours)', 'Charging Rate (kW)']:
    stat, p_val = stats.shapiro(df[feature].sample(min(500, len(df[feature]))))  # Sample for performance
    print(f"   {feature}: W = {stat:.3f}, p-value = {p_val:.4f}")
    normality = "Normal" if p_val > 0.05 else "Not Normal"
    print(f"     → Distribution is {normality}")
```

#### Step 2.2: Data Preprocessing & Encoding

```python
# Create a copy for preprocessing
df_processed = df.copy()

print("Data Preprocessing Steps:")
print("-" * 40)

# 1. Handle categorical variables using advanced encoding techniques
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

# Label encoding for ordinal categories
time_mapping = {'Night': 0, 'Morning': 1, 'Afternoon': 2, 'Evening': 3}
day_mapping = {'Monday': 0, 'Tuesday': 1, 'Wednesday': 2, 'Thursday': 3, 
               'Friday': 4, 'Saturday': 5, 'Sunday': 6}

df_processed['Time_of_Day_encoded'] = df_processed['Time of Day'].map(time_mapping)
df_processed['Day_of_Week_encoded'] = df_processed['Day of Week'].map(day_mapping)

# One-hot encoding for nominal categories
nominal_categoricals = ['Vehicle Model', 'Charging Station Location', 'Charger Type', 'User Type']
df_processed = pd.get_dummies(df_processed, columns=nominal_categoricals, prefix=nominal_categoricals)

print(f"1. Encoded categorical variables")
print(f"   Original shape: {df.shape}")
print(f"   After encoding: {df_processed.shape}")

# 2. Select features for clustering
# Remove identifiers and datetime columns, keep encoded versions
features_to_remove = ['User ID', 'Charging Station ID', 'Charging Station Location', 
                     'Charging Start Time', 'Charging End Time', 'Time of Day', 'Day of Week']

clustering_features = [col for col in df_processed.columns if col not in features_to_remove]
print(f"\n2. Selected {len(clustering_features)} features for clustering")

# Display the selected features
print("\nSelected features for clustering:")
for i, feature in enumerate(clustering_features, 1):
    print(f"   {i:2d}. {feature}")
```

#### Step 2.3: Normalization with Z-scores

```python
from sklearn.preprocessing import StandardScaler

# Separate the features for normalization
X_clustering = df_processed[clustering_features]

print("Feature Statistics Before Normalization:")
print("-" * 50)
print(X_clustering.describe().loc[['mean', 'std', 'min', 'max']].round(3))

# Apply Z-score normalization
scaler = StandardScaler()
X_normalized = scaler.fit_transform(X_clustering)
X_normalized_df = pd.DataFrame(X_normalized, columns=clustering_features, index=df_processed.index)

print("\nFeature Statistics After Z-score Normalization:")
print("-" * 55)
print(pd.DataFrame(X_normalized, columns=clustering_features).describe().loc[['mean', 'std', 'min', 'max']].round(3))

# Verify that means are ~0 and stds are ~1
print("\nVerification of Normalization:")
print(f"Mean of normalized features: {X_normalized.mean():.6f}")
print(f"Std of normalized features: {X_normalized.std():.6f}")

# Visualize the effect of normalization
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# Before normalization
axes[0].hist(X_clustering.iloc[:, 0], bins=30, alpha=0.7, color='lightblue', edgecolor='black')
axes[0].set_xlabel('Energy Consumed (kWh) - Original')
axes[0].set_ylabel('Frequency')
axes[0].set_title('Before Normalization')
axes[0].axvline(X_clustering.iloc[:, 0].mean(), color='red', linestyle='--', label=f'Mean: {X_clustering.iloc[:, 0].mean():.2f}')
axes[0].legend()

# After normalization
axes[1].hist(X_normalized[:, 0], bins=30, alpha=0.7, color='lightgreen', edgecolor='black')
axes[1].set_xlabel('Energy Consumed (kWh) - Normalized')
axes[1].set_ylabel('Frequency')
axes[1].set_title('After Z-score Normalization')
axes[1].axvline(X_normalized[:, 0].mean(), color='red', linestyle='--', label=f'Mean: {X_normalized[:, 0].mean():.2f}')
axes[1].legend()

plt.tight_layout()
plt.show()
```

### Phase 3: Dimensionality Reduction & Clustering Implementation
#### Step 3.1: Principal Component Analysis (PCA)

```python
from sklearn.decomposition import PCA
import numpy as np

print("PRINCIPAL COMPONENT ANALYSIS (PCA)")
print("=" * 50)

# Apply PCA to understand the intrinsic dimensionality
pca_full = PCA()
pca_full.fit(X_normalized)

# Calculate cumulative explained variance
explained_variance_ratio = pca_full.explained_variance_ratio_
cumulative_variance = np.cumsum(explained_variance_ratio)

# Plot explained variance
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.bar(range(1, len(explained_variance_ratio) + 1), explained_variance_ratio, alpha=0.7, color='skyblue', label='Individual')
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.title('Individual Explained Variance by PC')
plt.xticks(range(1, len(explained_variance_ratio) + 1, 2))

plt.subplot(1, 2, 2)
plt.plot(range(1, len(cumulative_variance) + 1), cumulative_variance, marker='o', linestyle='--', color='orange')
plt.axhline(y=0.95, color='r', linestyle='-', label='95% Variance')
plt.axhline(y=0.90, color='g', linestyle='-', label='90% Variance')
plt.axhline(y=0.85, color='b', linestyle='-', label='85% Variance')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Cumulative Explained Variance')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Print variance explained by top components
print("\nVariance Explained by Top Principal Components:")
for i, (var, cum_var) in enumerate(zip(explained_variance_ratio, cumulative_variance), 1):
    print(f"PC{i}: {var:.3f} ({cum_var:.3f} cumulative)")
    if cum_var >= 0.95:
        print(f"→ 95% variance reached at PC{i}")
        break

# Apply PCA for 2D and 3D visualization
pca_2d = PCA(n_components=2)
X_pca_2d = pca_2d.fit_transform(X_normalized)

pca_3d = PCA(n_components=3)
X_pca_3d = pca_3d.fit_transform(X_normalized)

print(f"\n2D PCA Shape: {X_pca_2d.shape}")
print(f"3D PCA Shape: {X_pca_3d.shape}")
print(f"Variance explained by 2 components: {pca_2d.explained_variance_ratio_.sum():.3f}")
print(f"Variance explained by 3 components: {pca_3d.explained_variance_ratio_.sum():.3f}")

# Analyze PCA component loadings
print("\nTop Features Contributing to Principal Components:")
feature_names = clustering_features
for i, component in enumerate(pca_2d.components_):
    print(f"\nPrincipal Component {i+1}:")
    # Get indices of top 5 absolute values
    top_indices = np.argsort(np.abs(component))[-5:][::-1]
    for idx in top_indices:
        print(f"  {feature_names[idx]}: {component[idx]:.3f}")
```

#### Step 3.2: Determining Optimal Number of Clusters

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score
import warnings
warnings.filterwarnings('ignore')

print("DETERMINING OPTIMAL NUMBER OF CLUSTERS")
print("=" * 55)

# Evaluate multiple metrics for different numbers of clusters
k_range = range(2, 11)
inertia = []
silhouette_scores = []
calinski_scores = []
davies_scores = []

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    cluster_labels = kmeans.fit_predict(X_normalized)
    
    inertia.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(X_normalized, cluster_labels))
    calinski_scores.append(calinski_harabasz_score(X_normalized, cluster_labels))
    davies_scores.append(davies_bouldin_score(X_normalized, cluster_labels))

# Plot the evaluation metrics
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# Elbow method (Inertia)
axes[0,0].plot(k_range, inertia, marker='o', linestyle='--', color='b')
axes[0,0].set_xlabel('Number of Clusters (k)')
axes[0,0].set_ylabel('Inertia (Within-cluster sum of squares)')
axes[0,0].set_title('Elbow Method for Optimal k')
axes[0,0].grid(True, alpha=0.3)

# Silhouette Score
axes[0,1].plot(k_range, silhouette_scores, marker='o', linestyle='--', color='g')
axes[0,1].set_xlabel('Number of Clusters (k)')
axes[0,1].set_ylabel('Silhouette Score')
axes[0,1].set_title('Silhouette Score vs Number of Clusters')
axes[0,1].grid(True, alpha=0.3)

# Calinski-Harabasz Score
axes[1,0].plot(k_range, calinski_scores, marker='o', linestyle='--', color='r')
axes[1,0].set_xlabel('Number of Clusters (k)')
axes[1,0].set_ylabel('Calinski-Harabasz Score')
axes[1,0].set_title('Calinski-Harabasz Score vs Number of Clusters')
axes[1,0].grid(True, alpha=0.3)

# Davies-Bouldin Score (lower is better)
axes[1,1].plot(k_range, davies_scores, marker='o', linestyle='--', color='purple')
axes[1,1].set_xlabel('Number of Clusters (k)')
axes[1,1].set_ylabel('Davies-Bouldin Score')
axes[1,1].set_title('Davies-Bouldin Score vs Number of Clusters\n(Lower is Better)')
axes[1,1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Print optimal k based on different metrics
optimal_k_silhouette = k_range[np.argmax(silhouette_scores)]
optimal_k_calinski = k_range[np.argmax(calinski_scores)]
optimal_k_davies = k_range[np.argmin(davies_scores)]

print(f"\nOptimal Number of Clusters based on:")
print(f"  Silhouette Score: k = {optimal_k_silhouette} (score: {max(silhouette_scores):.3f})")
print(f"  Calinski-Harabasz: k = {optimal_k_calinski} (score: {max(calinski_scores):.3f})")
print(f"  Davies-Bouldin: k = {optimal_k_davies} (score: {min(davies_scores):.3f})")

# Let's choose the optimal k (you can adjust this based on the metrics)
optimal_k = optimal_k_silhouette  # Using silhouette as primary metric
print(f"\nSelected optimal k = {optimal_k} for further analysis")
```

#### Step 3.3: Implementing Multiple Clustering Algorithms

```python
from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering
from sklearn.mixture import GaussianMixture
from sklearn.neighbors import NearestNeighbors

print("IMPLEMENTING MULTIPLE CLUSTERING ALGORITHMS")
print("=" * 55)

# Store results from different algorithms
clustering_results = {}

# 1. K-Means Clustering
print("1. K-MEANS CLUSTERING")
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
kmeans_labels = kmeans.fit_predict(X_normalized)
clustering_results['KMeans'] = kmeans_labels

kmeans_silhouette = silhouette_score(X_normalized, kmeans_labels)
print(f"   Silhouette Score: {kmeans_silhouette:.3f}")
print(f"   Cluster distribution: {np.bincount(kmeans_labels)}")

# 2. DBSCAN - Determining optimal epsilon
print("\n2. DBSCAN CLUSTERING")
# Find optimal epsilon using k-distance graph
neighbors = NearestNeighbors(n_neighbors=min(10, X_normalized.shape[0]//10))
neighbors_fit = neighbors.fit(X_normalized)
distances, indices = neighbors_fit.kneighbors(X_normalized)
distances = np.sort(distances[:, -1], axis=0)

plt.figure(figsize=(10, 6))
plt.plot(distances)
plt.xlabel('Data Points Sorted by Distance')
plt.ylabel('Epsilon (Distance to 10th Nearest Neighbor)')
plt.title('K-Distance Graph for DBSCAN Epsilon Selection')
plt.grid(True, alpha=0.3)
plt.show()

# Try DBSCAN with different parameters
dbscan_eps_values = [0.5, 1.0, 1.5, 2.0]
for eps in dbscan_eps_values:
    dbscan = DBSCAN(eps=eps, min_samples=10)
    dbscan_labels = dbscan.fit_predict(X_normalized)
    n_clusters = len(set(dbscan_labels)) - (1 if -1 in dbscan_labels else 0)
    n_noise = list(dbscan_labels).count(-1)
    
    if n_clusters > 0:
        sil_score = silhouette_score(X_normalized, dbscan_labels) if n_clusters > 1 else -1
        print(f"   eps={eps}: {n_clusters} clusters, {n_noise} noise points, silhouette: {sil_score:.3f}")
        if n_clusters >= 2 and n_clusters <= 10:  # Reasonable number of clusters
            clustering_results[f'DBSCAN_eps{eps}'] = dbscan_labels

# 3. Gaussian Mixture Models
print("\n3. GAUSSIAN MIXTURE MODELS")
gmm = GaussianMixture(n_components=optimal_k, random_state=42)
gmm_labels = gmm.fit_predict(X_normalized)
clustering_results['GMM'] = gmm_labels

gmm_silhouette = silhouette_score(X_normalized, gmm_labels)
print(f"   Silhouette Score: {gmm_silhouette:.3f}")
print(f"   Cluster distribution: {np.bincount(gmm_labels)}")

# 4. Agglomerative Hierarchical Clustering
print("\n4. AGGLOMERATIVE CLUSTERING")
agglo = AgglomerativeClustering(n_clusters=optimal_k)
agglo_labels = agglo.fit_predict(X_normalized)
clustering_results['Agglomerative'] = agglo_labels

agglo_silhouette = silhouette_score(X_normalized, agglo_labels)
print(f"   Silhouette Score: {agglo_silhouette:.3f}")
print(f"   Cluster distribution: {np.bincount(agglo_labels)}")
```

#### Step 3.4: Visualizing Clustering Results

```python
print("VISUALIZING CLUSTERING RESULTS")
print("=" * 50)

# Create subplots for different clustering algorithms
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
algorithms_to_plot = ['KMeans', 'GMM', 'Agglomerative', 'DBSCAN_eps1.5']

for i, (algo_name, ax) in enumerate(zip(algorithms_to_plot, axes.flat)):
    if algo_name in clustering_results:
        labels = clustering_results[algo_name]
        
        # Handle DBSCAN noise points
        if 'DBSCAN' in algo_name:
            scatter = ax.scatter(X_pca_2d[:, 0], X_pca_2d[:, 1], c=labels, 
                               cmap='tab10', alpha=0.7, s=50)
            n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
            n_noise = list(labels).count(-1)
            title = f'{algo_name}\n{n_clusters} clusters, {n_noise} noise'
        else:
            scatter = ax.scatter(X_pca_2d[:, 0], X_pca_2d[:, 1], c=labels, 
                               cmap='tab10', alpha=0.7, s=50)
            sil_score = silhouette_score(X_normalized, labels)
            title = f'{algo_name}\nSilhouette: {sil_score:.3f}'
        
        ax.set_xlabel('Principal Component 1')
        ax.set_ylabel('Principal Component 2')
        ax.set_title(title)
        ax.grid(True, alpha=0.3)
        
        # Add colorbar
        plt.colorbar(scatter, ax=ax)

plt.tight_layout()
plt.show()

# 3D visualization of the best performing algorithm
best_algo = max([(algo, silhouette_score(X_normalized, labels)) 
                for algo, labels in clustering_results.items() 
                if len(set(labels)) > 1], key=lambda x: x[1])

print(f"\nBest performing algorithm: {best_algo[0]} (Silhouette: {best_algo[1]:.3f})")

# Create 3D plot for the best algorithm
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

best_labels = clustering_results[best_algo[0]]
scatter = ax.scatter(X_pca_3d[:, 0], X_pca_3d[:, 1], X_pca_3d[:, 2], 
                    c=best_labels, cmap='tab10', alpha=0.7, s=50)

ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_zlabel('PC3')
ax.set_title(f'3D Cluster Visualization - {best_algo[0]}\n(Silhouette: {best_algo[1]:.3f})')

plt.colorbar(scatter)
plt.show()
```

#### Step 3.5: Comprehensive Cluster Evaluation

#### Step 4.4: Neural Network Approaches - Autoencoders for Clustering

```python
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, Dropout
from tensorflow.keras.optimizers import Adam

print("DEEP LEARNING APPROACH: AUTOENCODERS FOR CLUSTERING")
print("=" * 55)

# Build an autoencoder for representation learning
input_dim = X_normalized.shape[1]
encoding_dim = 8  # Compress to 8 dimensions

# Encoder
input_layer = Input(shape=(input_dim,))
encoded = Dense(32, activation='relu')(input_layer)
encoded = Dropout(0.2)(encoded)
encoded = Dense(16, activation='relu')(encoded)
encoded = Dropout(0.2)(encoded)
encoded = Dense(encoding_dim, activation='relu')(encoded)

# Decoder
decoded = Dense(16, activation='relu')(encoded)
decoded = Dropout(0.2)(decoded)
decoded = Dense(32, activation='relu')(decoded)
decoded = Dropout(0.2)(decoded)
decoded = Dense(input_dim, activation='linear')(decoded)

# Autoencoder model
autoencoder = Model(input_layer, decoded)
autoencoder.compile(optimizer=Adam(learning_rate=0.001), loss='mse')

print("Autoencoder Architecture:")
autoencoder.summary()

# Train the autoencoder
print("\nTraining Autoencoder...")
history = autoencoder.fit(X_normalized, X_normalized,
                         epochs=100,
                         batch_size=32,
                         shuffle=True,
                         validation_split=0.2,
                         verbose=0)

# Plot training history
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Mean Squared Error')
plt.title('Autoencoder Training History')
plt.legend()
plt.grid(True, alpha=0.3)

# Extract encoded representations
encoder = Model(input_layer, encoded)
X_encoded = encoder.predict(X_normalized)

print(f"Original dimension: {X_normalized.shape[1]}")
print(f"Encoded dimension: {X_encoded.shape[1]}")
print(f"Compression ratio: {X_normalized.shape[1]/X_encoded.shape[1]:.1f}x")

# Cluster in the encoded space
from sklearn.cluster import KMeans

kmeans_encoded = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
clusters_encoded = kmeans_encoded.fit_predict(X_encoded)

# Compare clustering performance
silhouette_original = silhouette_score(X_normalized, best_labels)
silhouette_encoded = silhouette_score(X_encoded, clusters_encoded)

plt.subplot(1, 2, 2)
comparison_data = [silhouette_original, silhouette_encoded]
labels = ['Original Features', 'Autoencoder Features']
colors = ['lightblue', 'lightgreen']
bars = plt.bar(labels, comparison_data, color=colors, alpha=0.7)
plt.ylabel('Silhouette Score')
plt.title('Clustering Performance: Original vs Autoencoder Features')
for bar, value in zip(bars, comparison_data):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.01, 
             f'{value:.3f}', ha='center', va='bottom')

plt.tight_layout()
plt.show()

print(f"\nClustering Performance Comparison:")
print(f"  Original features: Silhouette = {silhouette_original:.3f}")
print(f"  Autoencoder features: Silhouette = {silhouette_encoded:.3f}")
```
